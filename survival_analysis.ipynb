```python
# -*- coding: utf-8 -*-
"""
Created on Sat Jul 23 14:13:57 2022

@author: fayya
"""

from sksurv.kernels import clinical_kernel
from sksurv.svm import FastKernelSurvivalSVM
import numpy as np
import matplotlib.pyplot as plt
import pandas
import seaborn as sns
from sklearn.model_selection import ShuffleSplit, GridSearchCV, StratifiedShuffleSplit
import pandas as pd
from sksurv.datasets import load_veterans_lung_cancer
from sksurv.column import encode_categorical
from sksurv.metrics import concordance_index_censored
from sksurv.svm import FastSurvivalSVM
from sklearn.model_selection import train_test_split
from lifelines import KaplanMeierFitter
from lifelines.plotting import add_at_risk_counts
from lifelines.utils import concordance_index as cindex
from lifelines.statistics import logrank_test
from sklearn.model_selection import StratifiedShuffleSplit,StratifiedKFold
from tqdm import tqdm
from lifelines.plotting import add_at_risk_counts

idx = np.load(r'/content/drive/MyDrive/WSIMatrices/ShuffleNet_20x_512x512_noresize_1054patients/1054/slides_1054.npy') #slide IDs
df = pd.read_excel(r'/content/TCGA-BRCA-DX_CLINI (8).xlsx') #Surival Data
D = np.load(r'/content/drive/MyDrive/WSIMatrices/ShuffleNet_20x_512x512_noresize_1054patients/1054/blur10/shuffl_20x_512x512_new_blur_10.0.npy', mmap_mode='r') #MMD distance matrix

endpoint = 'DSS'
df = df[['PATIENT',endpoint+'_cr',endpoint+'_time_cr']]
df = df.rename(columns={endpoint+"_cr": "event", endpoint+"_time_cr": "time"})
df = df.set_index('PATIENT')
df = df.loc[idx]

K = np.exp(-1/np.median(D)*D)#formulate kernel
from sklearn.preprocessing import KernelCenterer

# construct data structure and exclude nans
y = np.vstack(((list(df.event==1.0),list(df.time)))).T
nidx = np.sum(np.isnan(y),axis=1)==0
K = K[nidx][:,nidx]
y = y[nidx]
EE,TT = y[:,0]>0, y[:,1]
count = 0
for item in EE:
  count += 1

"""
        dT = T_train[:, None] - T_train[None, :] #dT_ij = T_i-T_j
        dP = (dT>0)*E_train
"""
Tmax = 10*365
    
Y = [(EE[i],TT[i]) for i in range(len(y))]
Y = np.array(Y,dtype=[('Status', '?'), ('Survival_in_days', '<f8')])

# setup predictor
import warnings
warnings.filterwarnings("ignore", category=FutureWarning)#
param_grid = {'alpha': 2. ** np.arange(-4, 10, 1)}

def selectThreshold(Z,TT,EE,minprop = 0.1):
    thr_list = np.percentile(Z,np.linspace(minprop,1-minprop,int((1-minprop)*100))*100)
    V = [logrank_test(TT[Z>thr], TT[Z<=thr], EE[Z>thr],EE[Z<=thr]).test_statistic for thr in thr_list]
    thr = thr_list[np.argmax(V)]
    return thr
def surv_cutpointR(TT,EE,Z): 
    dfx = pd.DataFrame(data=np.vstack((TT,EE,Z)).T,columns=['time','event','Z'])
    import os

    from rpy2.robjects import pandas2ri, packages
    pandas2ri.activate()
    survminer = packages.importr('survminer')
    res = survminer.surv_cutpoint(dfx,time = 'time',event = 'event',variables = 'Z') #function call
    d = { key : res.rx2(key)[0] for key in res.names } #https://stackoverflow.com/questions/24152160/converting-an-rpy2-listvector-to-a-python-dictionary
    thr = d['cutpoint'][0]
    return thr

from sksurv.metrics import (
    concordance_index_censored,
    concordance_index_ipcw,
    cumulative_dynamic_auc,
    integrated_brier_score,
)
def plot_cumulative_dynamic_auc(risk_score, y_train, y_test,times):
    auc, mean_auc = cumulative_dynamic_auc(y_train, y_test, risk_score, times)
    if np.isnan(mean_auc ): mean_auc = np.nanmean(auc)
    plt.plot(times,[mean_auc]*len(times),  linestyle="--")
    plt.plot(times, auc, marker="o")
    plt.xlabel("Timeline")
    plt.ylabel("AUC")
    return auc,mean_auc

rng = np.random.RandomState()
times = np.linspace(0.5,10)*365
EE = np.asarray(EE,dtype=int)
TT = np.asarray(TT,dtype=int)
for alpha in param_grid['alpha']:
    alpha = 0.125
    C = []
    P = []
    A=[]

    #out-of-bag bootstrap.
    run_counter = 0
    for run in tqdm(range(50)):#run,(index_train, index_test) in tqdm(enumerate(sss.split(K, EE))):
        index_train = list(rng.choice(np.nonzero(EE==0)[0],size = len(EE)-np.sum(EE),replace=True))+list(rng.choice(np.nonzero(EE==1)[0],size = np.sum(EE),replace=True))
        index_test = list(set(range(len(EE))).difference(index_train))

        #stratified out of bag bootstrap

        # Creating the X, T and E input
        K_train = K[index_train][:,index_train]
        y_train,y_test = Y[index_train], Y[index_test]
        T_train, T_test = TT[index_train], TT[index_test]
        E_train, E_test = EE[index_train], EE[index_test]
        
        kssvm = FastKernelSurvivalSVM(kernel="precomputed", alpha = alpha, max_iter  = 2000)# random_state=0)
        kssvm = kssvm.fit(K_train,y_train)
        
        Z = kssvm.predict(K[:,index_train])
        Z_train,Z_test = Z[index_train], Z[index_test]

        # censor:
        tidx = T_train>Tmax
        T_train[tidx]=Tmax
        E_train[tidx]=False
        tidx = T_test>Tmax
        T_test[tidx]=Tmax
        E_test[tidx]=False

        thr = selectThreshold(Z_train,T_train,E_train)

        results = logrank_test(T_test[Z_test>thr], T_test[Z_test<=thr], E_test[Z_test>thr],E_test[Z_test<=thr])
        p_tt = results.p_value
        cc_ttx = cindex(T_test, -Z_test, E_test)
        weight = 0#0.632
        cc_ttx = weight * cindex(T_train, -Z_train, E_train) + (1.0 - weight) * cindex(T_test, -Z_test, E_test)
        C.append(cc_ttx);P.append(p_tt)
 
        #Plot Kaplan Meier curve
        plt.figure()
        ax = plt.gca()            
            
        test_low = KaplanMeierFitter().fit(T_test[Z_test<=thr], event_observed=E_test[Z_test<=thr], label = 'L')#NelsonAalenFitter()
        test_high = KaplanMeierFitter().fit(T_test[Z_test>thr], event_observed=E_test[Z_test>thr], label = 'H')#NelsonAalenFitter()

        ax = test_high.plot_survival_function(ax=ax, show_censors = False)
        ax = test_low.plot_survival_function(ax=ax, show_censors = False)
        tt = ax.get_xticklabels()

        plt.title('logrank p-value: %0.3f c-index: %0.3f' %(p_tt,cc_ttx));
          
        plt.ylim(0.5, 1)
        plt.legend(loc='upper right')
        plt.savefig('/content/runs/'+str(run_counter) +'.png')                                
        plt.show()
        run_counter += 1

    print('Alpha',alpha,'C-index',np.mean(C),np.std(C),'P-value',2*np.median(P))
    np.save("C.npy",C)
    np.save("P.npy",P)
    break
```
